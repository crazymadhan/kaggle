```python
X = Bank_Loan.drop(['Personal Loan','ID','ZIP Code'],axis=1)
Y = Bank_Loan['Personal Loan']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
SC = StandardScaler()
X_train = SC.fit_transform(X_train)
X_test = SC.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
def predict(model,X_train,Y_train,X_test,Y_test):
    Y_Pred_train = model.predict(X_train)
    Y_pred_test = model.predict(X_test)
    print(metrics.confusion_matrix(Y_train,Y_Pred_train))
    print(metrics.confusion_matrix(Y_test,Y_pred_test))
```

```python
Log_Reg = LogisticRegression()
Log_Reg.fit(X_train,Y_train)
predict(Log_Reg,X_train,Y_train,X_test,Y_test)
```

```python
GNB = GaussianNB()
GNB.fit(X_train,Y_train)
predict(GNB,X_train,Y_train,X_test,Y_test)
```

```python
KNN = KNeighborsClassifier()
KNN.fit(X_train,Y_train)
predict(KNN,X_train,Y_train,X_test,Y_test)
```

```python
# Logistic regression got the best result in training and test data .
# The features are not in gaussian distribution and so Gaussian naive bayes doesnt work very well here.
# The features are unable to classify the data with respect to its neighbours as most of the data point are mingled with each others classification and so KNN also doesnt give better result.
# Logistic regression result is little better than these two as it was able to find a line for the features to split the classification. In comparision to the other two it is better but it is not the best
# In this data there are not much positive or negative correlation most of the data has to be split in vertical and horizontal so these algo's are unable to perform very well
# Decision tree algo's will work very well with these kind of distribution.
```
